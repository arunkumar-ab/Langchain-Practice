```python
#Web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader(web_path=("https://en.wikipedia.org/wiki/Surya"),
                       bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                           class_=('vector-body')
                       ))
                       )
docs = loader.load()
```
